Import drive and mount drive to grab datasets
Import pandas for reading and writing csv

```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
```

Grab datasets

```python
df1=pd.read_csv('/content/drive/MyDrive/PHYS 549/GW1data.csv')
df2=pd.read_csv('/content/drive/MyDrive/PHYS 549/GW2data.csv')
df3=pd.read_csv('/content/drive/MyDrive/PHYS 549/GW3data.csv')
dfloner=pd.read_csv('/content/drive/MyDrive/PHYS 549/GW200308.csv')
print(df1)
print(df2)
print(df3)
print(dfloner)
```

Creat dataframe of all data

```python
dftotal=pd.concat([df1,df2,df3,dfloner])
print(dftotal)
```

Save csv of all data

```python
dftotal.to_csv('/content/drive/MyDrive/PHYS 549/totalgwdata.csv')
```

Lets look at all data so import seaborn and matplotlib

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

plot distribution of all gw events

```python
sb.color_palette("mako", as_cmap=True)
sb.kdeplot(data=dftotal, x='dec',y='ra',weights='weights',Fill=True)
plt.show()
```
